**Introduction to Natural Language Processing**

Natural Language Processing (NLP) is a branch of artificial intelligence that focuses on the interaction between computers and human languages. The goal of NLP is to enable computers to understand, interpret, and respond to human language in a way that is meaningful and useful. NLP combines linguistics, computer science, and machine learning to process and analyze large amounts of natural language data.

**Here are some of the most popular Python libraries for Natural Language Processing (NLP):**

**NLTK (Natural Language Toolkit):**

NLTK is one of the most popular and powerful libraries for working with human language data (text).

Provides a wide range of tools for text processing tasks such as tokenization, stemming, tagging, parsing, and semantic reasoning.

It also comes with a large collection of datasets and text corpora.

**spaCy:**

SpaCy is a fast and efficient NLP library designed for production use cases.

It supports deep learning integration, has pre-trained models for various languages, and handles tasks like tokenization, named entity recognition (NER), part-of-speech tagging, dependency parsing, and more.

**SpaCy provides several pre-trained language models. The most commonly used models include:**

    en_core_web_sm for English (small model)

    en_core_web_md for English (medium model)

    en_core_web_lg for English (large model)

    xx_ent_wiki_sm for multi-language model

**TextBlob:**

TextBlob is a simple library for NLP tasks such as part-of-speech tagging, noun phrase extraction, sentiment analysis, and text translation.

It is built on top of NLTK and the Pattern library.

```python
import requests
import pandas as pd

# URL of the raw CSV file
url = "https://raw.githubusercontent.com/omaarelsherif/Movie-Reviews-Sentiment-Analysis-Using-Machine-Learning/main/Dataset/IMDB.csv"

# Fetch the content from the URL
response = requests.get(url)
response.raise_for_status()  # Check if the request was successful

# Save the content to a CSV file
csv_file_path = "C:/MAFAS/APU/UNIVERSITY COLLABORATIONS/VET Institute of Arts and Science/12-17 September 2025/Workshop/DAY 01/IMDb_Reviews.csv"
with open(csv_file_path, 'wb') as file:
    file.write(response.content)

print(f"Data has been saved to {csv_file_path}")
```
